```python
import pandas as pd
```

```python
#load the csv file
df_exams=pd.read_csv("sample_data/StudentsPerformance.csv")
```

```python
df_exams.head()
```

**1 Value Counts**

```python
#counting gender elements

#len function(python inbuild function)
len(df_exams['gender'])

#Another way we can use the count() in pandas
df_exams['gender'].count()
```

```python
#what if we want to count the males or females from our category
df_exams['gender'].value_counts()     #make sure the column has only two values
```

```python
#with the value_counts() we can do more with it e.g return the relative frequency(divide all values by the sum of values)
df_exams['gender'].value_counts(normalize=True)    #this is the %
```

```python
#Another example
#count the elements by category
df_exams['parental level of education'].value_counts()
```

```python
#return the relative frequency and round to 2 decimal
df_exams['parental level of education'].value_counts(normalize=True).round(2)
```
