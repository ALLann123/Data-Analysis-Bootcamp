**1. Create a dataframe from an array**


***1.1 Option 1 ***

```python
#create an array i.e use numpy
data=np.array([[1,4], [2,5],[3,7]])    #each pair of the list represent rows
```

```python
#create a data frame with pandas
df=pd.DataFrame(data, index=['row1', 'row2', 'row3'], columns= ['col1', 'col2'])

```

```python
#show the dataframe
df
```

**2. Creating a DataFrame from a dictionary**


```python
#lists used for this example
states=['Nairobi', 'Nakuru', 'Transnzoia']    #this is our first list
population=[4397073, 2162202, 990341]         #this is our second list


```

```python
#now lets store the list above in a dictionary
dict_states={'States':states, 'Population':population}    #use the lists above to create a dictionary
```

```python
#now lets create a dataframe
df_population=pd.DataFrame(dict_states)   #now use dictionary above and load it in pandas
```

```python
#display our dataframe
df_population
```
