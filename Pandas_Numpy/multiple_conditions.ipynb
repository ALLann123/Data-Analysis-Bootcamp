***1 Creating a conditional Column from more than 2 choices:np.select()***

```python
import numpy as np
```

```python
#create an array based on multiple price tiers (+2 choices)

#create "conditions" and "values" variables
conditions=[
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']>800) & (df_laptops['Price_euros']<=2000),
    df_laptops['Price_euros']<=800,
]


#the position of each element of the conditions list shoult correspond directly to the value list below
values=['Too Expensive','Expensive','Affordable', 'Cheap']
```

```python
#set it to a new column. The last parameter is the default
df_laptops['Price_tiers']=np.select(conditions, values, default='Unknown')
```

```python
#show column
df_laptops.head(10)
```

```python
#lets check the range number
df_laptops.value_counts('Price_tiers')
```

```python
#lets display the too expensive ones here
df_laptops[df_laptops['Price_tiers']=="Too Expensive"]
```

```python
#create an array based on multiple screen size tiers
screen_conditions=[
    df_laptops['Inches']>16,
    (df_laptops['Inches']>14) &(df_laptops['Inches']<=16),
    (df_laptops['Inches']>12) & (df_laptops['Inches']<=14),
    df_laptops['Inches']<=12
]

screen_values=['Too Big', 'Big', 'Small', 'Too Small']
```

```python
np.select(screen_conditions, screen_values, default="Unknown")
```

```python
#set it in a new column
df_laptops['Screen_tier']=np.select(screen_conditions, screen_values, default="Unknown")
```

```python
#count values inside "screen size/screen_tier"column
df_laptops.value_counts('Screen_tier')
```
