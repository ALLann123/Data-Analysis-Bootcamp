***1.2 Adding a new column with an array***

```python
#we know the dataframe has 1000 row.
df_exams.index
```

```python
#we create an array with 1000 elements as the dataframe has an index upto 1000
import numpy as np
```

```python
#create an array of 1000 elements
language_score=np.arange(0,1000)   #this are the limits of my dataframe i.e starting index and the last one
```

```python
#verify the length
len(language_score)
```

```python
#add the new column to the dataframe with an array
df_exams['language score'] = language_score
df_exams
```

```python
#Now lets give the marks to our new column using numpy and random
#between 1-100
int_language_score=np.random.randint(1, 100, size=1000)
```

```python
#check the minimum value of the array
min(int_language_score)
```

```python
#check the maximum value
max(int_language_score)
```

```python
#add our new column to the dataframe
df_exams['language score'] = int_language_score
df_exams
```

```python
#lets create random float numbers between 32 and 38
stud_temp=np.random.uniform(32, 38, size=1000)
```

```python
#lets add our new floating point column to the table
df_exams['Temperature°C'] = stud_temp
#display
df_exams
```
