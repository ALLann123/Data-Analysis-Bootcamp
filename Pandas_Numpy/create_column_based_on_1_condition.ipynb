**1. Create a column based on 1 conditions: np.where()**

```python
import numpy as np
```

```python
#create an array based on price tiers

#the first argument inthe np.where is the condition followed by value of condition when true then finally the value of condition when false
np.where(df_laptops['Price_euros']>2000, 'Expensive','Cheap')


```

```python
#now lets set our array above to a new column
df_laptops['Price_tier']=np.where(df_laptops['Price_euros']>2000, 'Expensive','Cheap')
```

```python
df_laptops.head(5)
```

```python
#count values in the dataframe
df_laptops.value_counts('Price_tier')
```

```python
#display 5 cheap laptops only
cheap_laptops=df_laptops[df_laptops['Price_tier']=="Cheap"]
cheap_laptops.head(5)
```

```python
#display the 5 most cheapest laptops
cheap_laptops.sort_values(by='Price_euros', ascending=True)
```
