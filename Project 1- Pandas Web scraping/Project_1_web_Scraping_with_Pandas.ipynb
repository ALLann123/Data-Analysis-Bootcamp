**2 Read .csv from Multiple URLs with Pandas**

***Link: root+season+league***
The links have a pattern:


*  https://www.football-data.co.uk/mmz4281/2425/E0.csv
*   https://www.football-data.co.uk/mmz4281/2425/E1.csv



```python
#;ink structure
"https://www.football-data.co.uk/mmz4281/"+"2425"+"/"+"E1"+".csv"
```

```python
#We are going to extract the csv files using a for loop
#create a root variable
root="https://www.football-data.co.uk/mmz4281/"
```

```python
#create list of leagues
leagues=['E0', 'E1', 'E2', 'E3', 'EC']
frames=[]  #create an empty list to store the dataframe

#lets loop throught the list above
for league in leagues:
  df=pd.read_csv(root+"2425"+"/"+league+".csv")
  frames.append(df)     #now add each dataframe to our empty list
```

```python
#length of frames
len(frames)   #we are supposed to get 5 elements in our list
```

```python
#show 1st, 2nd and 3rd element
frames[0]  #the first element of our list
```

```python
frames[4]
```

**2.2 Multiple Seasons**

```python
for season in range(19, 25):
  print(str(season)+ str(season+1))
```

```python
#loop through the range of seasons and insert new column
leagues=['E0', 'E2', 'E3']
frames=[]  #create an empty list to store the dataframe

#lets loop throught the list above
for league in leagues:
  for season in range(19, 25):
    df=pd.read_csv(root+str(season)+ str(season+1)+"/"+league+".csv")
    df.insert(1, 'Season',season)
    frames.append(df)     #now add each dataframe to our empty list
```

```python
#total frames( 6 seasons x 4 leagues)
len(frames)
```

```python
#show 1st and last element
frames[17]
```

**3 Organize All the Data in a Dictionary**

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1',
              'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1',
              'English Premier League':'E0',
              'English League 1':'E2',
              'English League 2':'E3',
             }
```

```python
dict_countries['English League 1']
```

```python
#lets see what we get when we loop through a dictionary
for league in dict_countries:
  print(dict_countries[league])  #this allows us to get the values in a dictionary
```

```python
dict_historical_data={}

#loop through the range of seasons and insert new column
#leagues=['E0', 'E2', 'E3']
 #create an empty list to store the dataframe

#lets loop throught the list above
for league in dict_countries:
  frames=[] #prevents accumilation of the previous data and emptys the list with each iteration

  for season in range(19, 25):
    df=pd.read_csv(root+str(season)+ str(season+1)+"/"+dict_countries[league]+".csv")
    df.insert(1, 'Season',season)
    frames.append(df)     #now add each dataframe to our empty list

  df_concat=pd.concat(frames)
  dict_historical_data[league]=df_concat  #assign name of league and its dataframe

```
